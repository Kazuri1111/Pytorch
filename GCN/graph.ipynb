```python
import matplotlib.pyplot as plt
with open("./HOMO", "r") as f:
    line = f.readlines()
    line.pop(0)
train_loss = []
valid_loss = []
for l in line:
    l = l.split(",")
    train = float(l[0].split(":")[1])
    valid = float(l[1].split(":")[1])
    train_loss.append(train)
    valid_loss.append(valid)
plt.plot([i for i in range(len(train_loss))],train_loss)
plt.plot([i for i in range(len(valid_loss))],valid_loss)
```
