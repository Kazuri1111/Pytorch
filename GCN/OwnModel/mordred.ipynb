```python
from rdkit import Chem

def read_smiles(smiles_str):
    mol = Chem.MolFromSmiles(smiles_str)
    return mol
    
from mordred import Calculator, descriptors

def calculate_mordred(mol):
    calc = Calculator(descriptors, ignore_3D=True)
    mordred_descriptors = calc(mol)
    return mordred_descriptors
    
from rdkit.Chem import AllChem

def calculate_ecfp(mol, radius=2, nBits=1024):
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
    return list(ecfp)
    
from torch_geometric.data import Data

def calculate_gnn(mol):
    # ここでは簡略化のためGNNのダミーコードを示します。
    # 実際にはGNNモデルを設定し、分子グラフからGNN記述子を計算します。
    return Data(x=None, edge_index=None)
    
def combine_descriptors(mordred_desc, ecfp_desc, gnn_desc):
    return mordred_desc + ecfp_desc + gnn_desc
    
if __name__ == "__main__":
    smiles_str = "CCO"
    mol = read_smiles(smiles_str)
    
    mordred_desc = calculate_mordred(mol)
    ecfp_desc = calculate_ecfp(mol)
    gnn_desc = calculate_gnn(mol)
    
    combined_desc = combine_descriptors(mordred_desc, ecfp_desc, gnn_desc)
```
